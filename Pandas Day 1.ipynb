```python
#merge all the facbook likes columns into 1 one columns 'all fb likes'
movies["all fb likes"] = (movies["director_facebook_likes"]+movies['actor_3_facebook_likes']+movies["actor_1_facebook_likes"]+movies["actor_2_facebook_likes"])
movies["all fb likes"].head()
```

```python
#merge all actor likes into one column 
movies["all actors likes"] = (movies["actor_1_facebook_likes"]+movies["actor_2_facebook_likes"]+movies["actor_3_facebook_likes"])
```

```python
#making all null values of 'all actors likes' to zero
movies["all actors likes"].fillna(0).head()
```

```python
#Now checking null values of all fb likesis equal to null values of another column 'cast_total_facebook_likes' 
movies["all fb likes"].isna().sum() <= movies["cast_total_facebook_likes"].isna().sum() 
#So, "cast_total_facebook_likes" has less null values than "all fb likes"
```

```python
#Now let's check whether "cast_total_facebook_likes" is from director likes or actor likes
movies["cast_total_facebook_likes"].isna().sum() == movies["all actors likes"].isna().sum()
#It seems that cast_total_facebook_likes has more null values than "all actor likes"
```

```python
#Now let's calc what percentage of cast_total_facebook_likes is of 'all actors likes'
movies["pct_actor_likes"] =  movies["all actors likes"] / movies["cast_total_facebook_likes"]
movies["pct_actor_likes"].head()
```
